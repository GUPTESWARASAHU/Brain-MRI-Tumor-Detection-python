```python
# Full pipeline for Brain MRI Tumor Detection (Colab-ready)
# 1) Upload kaggle.json to Colab Files sidebar before running
# 2) Use GPU runtime

# ----------------------------
# Install & Imports
# ----------------------------
!pip install -q kaggle python-docx

import os, glob, zipfile, random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, GlobalAveragePooling2D, Input
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

print("TF version:", tf.__version__)

# ----------------------------
# Kaggle dataset download
# ----------------------------
# Make sure kaggle.json is uploaded into the Colab working dir (Files sidebar)
KAGGLE_PATH = '/root/.kaggle'
os.makedirs(KAGGLE_PATH, exist_ok=True)
if os.path.exists('kaggle.json'):
    !cp kaggle.json /root/.kaggle/
else:
    if not os.path.exists('/root/.kaggle/kaggle.json'):
        raise FileNotFoundError("Please upload 'kaggle.json' to Colab (Files sidebar) and re-run this cell.")
!chmod 600 /root/.kaggle/kaggle.json

!kaggle datasets download -d navoneel/brain-mri-images-for-brain-tumor-detection -q
!unzip -q brain-mri-images-for-brain-tumor-detection.zip -d dataset
print("Dataset downloaded to ./dataset")

# ----------------------------
# EDA: check image counts + sample display
# ----------------------------
tumor_paths = glob.glob('dataset/yes/*.jpg') + glob.glob('dataset/Yes/*.jpg')
no_paths = glob.glob('dataset/no/*.jpg') + glob.glob('dataset/No/*.jpg')
print("Tumor images:", len(tumor_paths))
print("No tumor images:", len(no_paths))

# Display a few samples if available
sample_paths = (tumor_paths[:3] if len(tumor_paths)>0 else []) + (no_paths[:3] if len(no_paths)>0 else [])
if sample_paths:
    plt.figure(figsize=(10,4))
    for i,p in enumerate(sample_paths):
        img = plt.imread(p)
        plt.subplot(1, len(sample_paths), i+1)
        plt.imshow(img)
        plt.title("Tumor" if i<3 else "No Tumor")
        plt.axis('off')
    plt.show()

# ----------------------------
# Preprocessing: safe loader (case-insensitive), resize, convert BGR->RGB, normalize
# ----------------------------
IMG_SIZE = 150
data, labels = [], []
root_dir = 'dataset'
if not os.path.exists(root_dir):
    raise FileNotFoundError("dataset directory missing. Run Kaggle download step above.")

found_folders = [f for f in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, f))]
lower_map = {f.lower(): f for f in found_folders}
if 'yes' in lower_map and 'no' in lower_map:
    dirs = [lower_map['no'], lower_map['yes']]
else:
    raise FileNotFoundError("Couldn't find 'yes'/'no' folders in dataset/")

for label, folder in enumerate(dirs):  # label=0 -> no, label=1 -> yes
    folder_path = os.path.join(root_dir, folder)
    for fname in os.listdir(folder_path):
        path = os.path.join(folder_path, fname)
        try:
            img = cv2.imread(path)
            if img is None:
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            data.append(img)
            labels.append(label)
        except Exception as e:
            print("Skipping", path, ":", e)

data = np.array(data, dtype='float32') / 255.0
labels = np.array(labels, dtype='int32')
print("Total images loaded:", len(data))
print("Class counts (no=0, yes=1):", np.bincount(labels))

# ----------------------------
# Train-test split (stratified)
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    data, labels, test_size=0.20, random_state=42, stratify=labels
)
print("Shapes -> X_train:", X_train.shape, "X_test:", X_test.shape)

# ----------------------------
# Data Augmentation
# ----------------------------
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest'
)
datagen.fit(X_train)

# ----------------------------
# Callbacks
# ----------------------------
os.makedirs('models', exist_ok=True)
cb_early = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
cb_cnn_ckpt = ModelCheckpoint('models/best_custom_cnn.h5', monitor='val_loss', save_best_only=True)
cb_resnet_ckpt = ModelCheckpoint('models/best_resnet.h5', monitor='val_loss', save_best_only=True)

# ----------------------------
# Custom CNN
# ----------------------------
cnn = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
cnn.compile(optimizer=Adam(learning_rate=1e-3), loss='binary_crossentropy', metrics=['accuracy'])
print(cnn.summary())

history_cnn = cnn.fit(
    datagen.flow(X_train, y_train, batch_size=32),
    validation_data=(X_test, y_test),
    epochs=12,
    callbacks=[cb_early, cb_cnn_ckpt],
    verbose=1
)

# Save final
cnn.save('models/final_custom_cnn.h5')

# ----------------------------
# ResNet50 Transfer Learning
# ----------------------------
base = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base.trainable = False  # freeze base

x = base.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
out = Dense(1, activation='sigmoid')(x)
resnet_model = Model(inputs=base.input, outputs=out)

resnet_model.compile(optimizer=Adam(1e-4), loss='binary_crossentropy', metrics=['accuracy'])
print(resnet_model.summary())

history_resnet = resnet_model.fit(
    datagen.flow(X_train, y_train, batch_size=16),
    validation_data=(X_test, y_test),
    epochs=8,
    callbacks=[cb_early, cb_resnet_ckpt],
    verbose=1
)

resnet_model.save('models/final_resnet.h5')

# Optionally fine-tune some layers:
# unfreeze last conv block and fine-tune for a few epochs (if you have enough GPU/time)
# base.trainable = True
# for layer in base.layers[:-30]:
#     layer.trainable = False
# resnet_model.compile(optimizer=Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy'])
# history_finetune = resnet_model.fit(...)

# ----------------------------
# Evaluation (use best ResNet model if checkpointed)
# ----------------------------
# load best if exists
if os.path.exists('models/best_resnet.h5'):
    model_to_eval = load_model('models/best_resnet.h5')
else:
    model_to_eval = resnet_model

y_prob = model_to_eval.predict(X_test, verbose=0)
y_pred = (y_prob > 0.5).astype(int).flatten()

print("Classification report (ResNet):")
print(classification_report(y_test, y_pred, target_names=['No Tumor','Tumor']))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix'); plt.show()

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
plt.figure(figsize=(6,5))
plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.3f}')
plt.plot([0,1],[0,1],'--', alpha=0.5)
plt.xlabel('False Positive Rate'); plt.ylabel('True Positive Rate'); plt.title('ROC Curve'); plt.legend(); plt.show()

# ----------------------------
# Visualize some predictions
# ----------------------------
import numpy as np
n = min(6, len(X_test))
idxs = np.random.choice(len(X_test), n, replace=False)
plt.figure(figsize=(10,6))
for i, idx in enumerate(idxs):
    plt.subplot(2,3,i+1)
    plt.imshow(X_test[idx])
    plt.title(f"True:{y_test[idx]} Pred:{int(y_pred[idx])}")
    plt.axis('off')
plt.tight_layout(); plt.show()

# ----------------------------
# Save training history arrays for plotting outside notebook if needed
# ----------------------------
import pickle
with open('models/history_cnn.pkl','wb') as f:
    pickle.dump(history_cnn.history, f)
with open('models/history_resnet.pkl','wb') as f:
    pickle.dump(history_resnet.history, f)

print("All done. Models and histories saved in ./models")

```
